```python
import pandas as pd
data = {
    'Region': ['East','East','West','West','East', 'West'],
    'Product': ['Apple','Orange','Apple','Orange', 'Apple', 'Orange'],
    'Sales': [100,150,200,130, 90, 160],
    'Quantity': [10,15,20,13, 9, 16]
}

df = pd.DataFrame(data)
df
```

```python
pivot_table = pd.pivot_table(data = df, index='Region', columns='Product', values='Sales', aggfunc='sum')
pivot_table
```

```python
pivot_table = pd.pivot_table(data = df, index='Region', columns='Product', values='Sales', aggfunc=['sum','mean','count'])
pivot_table
```

```python
pivot_table = pd.pivot_table(data = df, index='Region', columns='Product', values=['Sales','Quantity'], aggfunc=['sum','mean','count'])
pivot_table
```
